```python
from pathlib import Path
import json
import subprocess

MODEL_NAME = 'Qwen/Qwen3.5-4B'
DEVICE = 'cuda'
MAX_LENGTH = 160
PROFILES = ['short', 'medium', 'long']

CONCEPT_JSON = Path('/content/ABPT/archive/qwen35_4b_anchor_concept_direction_map.json')
CONCEPT_MD = Path('/content/ABPT/docs/research/qwen35_4b_anchor_concept_direction_map.md')
CARRY_JSON = Path('/content/ABPT/archive/qwen35_4b_anchor_carryover_probe.json')
CARRY_MD = Path('/content/ABPT/docs/research/qwen35_4b_anchor_carryover_probe.md')
```

```python
cmd = [
    'python', 'scripts/run_qwen_anchor_concept_direction_map.py',
    '--model', MODEL_NAME,
    '--device', DEVICE,
    '--max_length', str(MAX_LENGTH),
    '--profiles', *PROFILES,
    '--output_json', str(CONCEPT_JSON),
    '--output_md', str(CONCEPT_MD),
]
print('RUN:', ' '.join(cmd))
result = subprocess.run(cmd, text=True, capture_output=True)
print(result.stdout)
if result.stderr:
    print(result.stderr)
result.check_returncode()
```

```python
cmd = [
    'python', 'scripts/run_qwen_anchor_carryover_probe.py',
    '--model', MODEL_NAME,
    '--device', DEVICE,
    '--max_length', str(MAX_LENGTH),
    '--profiles', *PROFILES,
    '--output_json', str(CARRY_JSON),
    '--output_md', str(CARRY_MD),
]
print('RUN:', ' '.join(cmd))
result = subprocess.run(cmd, text=True, capture_output=True)
print(result.stdout)
if result.stderr:
    print(result.stderr)
result.check_returncode()
```

```python
for path in (CONCEPT_JSON, CARRY_JSON):
    payload = json.loads(path.read_text(encoding='utf-8'))
    print('FILE:', path.name)
    print(json.dumps(payload['metadata'], ensure_ascii=False, indent=2))
    print()
```

```python
for path in (CONCEPT_MD, CARRY_MD):
    print('=' * 80)
    print(path.name)
    print('=' * 80)
    text = path.read_text(encoding='utf-8')
    print('\n'.join(text.splitlines()[:80]))
    print()
```
